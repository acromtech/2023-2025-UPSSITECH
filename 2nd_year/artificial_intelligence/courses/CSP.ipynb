### Enoncé 30 : Configuration produit
* Un processeur `p`
* De la mémoire vive `m`
* Un disque dur `d`

On a le choix entre
* 3 type de `p` : `p1`, `p2`, `p3`
* 4 type de `m` : `m1`, `m2`, `m3`, `m4` 
* 3 type de `d` : `d1`, `d2`, `d3`, `d4`

L'indice indique l'année de sortie (plus il est grand plus il est récent)

**Contraintes**
* `d1` et `p1` incompatible
* `pi` = `di` avec i l'indice
* Si `p2`+`m2` -> `d2`

**1. Quel choix faites vous pour les variables ?**  
Les variables sont `V={p,m,d}`

**2. Quels sont les domaines des variables ?**  
Le domaine est donc de la forme `D{Dp,Dm,Dp}` avec 
* `Dp={1,2,3}`
* `Dm={1,2,3,4}`
* `Dd={1,2,3}`

**3. Quelles sont les contraintes qui portent sur ces variables ?**  
On fera les tuple interdit (et non les tuple autorisé mais aussi possible) `C={Cpd U Cpm U Cpmd}` avec
* `Cpd={(1,3)}`
* `Cpm={(2,1),(3,1),(3,2)}`
* `Cpmd={(2,2,1),(2,2,3),...}`
