### (b) le code source correspondant

```
#include <stdio.h>
#include <math.h>

void main(){
  int nbl,nbc,pxlc=0;
  int niveau;
  
  scanf("%d",&niveau);
  
  if(niveau<256){
    if((int)log2(niveau)==log2(niveau)){
      
      scanf("%d",&nbl);
      scanf("%d",&nbc);
      
      for(int i=0;i<nbl;i++){
        for(int j=0;j<nbc;j++){
          scanf("%d",&pxlc);
          printf("%5d",((pxlc*niveau)/256));
        }
        printf("\n\r");
      }
    }
    else printf("Erreur 1 : le niveau de gris est supérieur à 256");
  }
  else printf("Erreur 2 : le niveau de gris n’est pas une puissance de 2");
}
```

### (c) le contenu du Makefile de la question 21

**Question 21 :**

Contenu du makefile :

```
mon_module_image.out: mon_module_image.o main.o
    gcc -o mon_module_image.out mon_module_image.o main.o
mon_module_image.o: ferrane_module_image.c 
    gcc -c ferrane_module_image.c -o mon_module_image.o  -W -Wall -ansi -pedantic
main.o: main.c ferrane_module_image.h
    gcc -o main.o -c main.c -W -Wall -ansi -pedantic
clean:
    rm -rf *.o
```

- **Le "Makefile" contient un script qui suit le procesus suivant :**
    - Les dépendances "main.c" et "ferrane_module_image.h" sont rangées dans la cible intermédiaire "main.o" puis la compile
    - La dépendance "ferrane_module_image.c" est rangé dans la cible intermédiaire "mon_module_image.o" puis la compile
    - Les dépendances "mon_module_image.o" "main.o" sont rangées dans la cible finale "mon_module_image.out" puis la compile
    - (Par la suite les fichiers intermédiaires .o pourront être supprimés en utilisant la commande "make clean")

- **Les options utilisées :**
    - `-W`,`-Wall` et `-pedantic` sont des options relatives aux "warning". Celles-ci activent tous les "warning" relatif au fichier compilé et conseille l'utilisateur afin de parrer une éventuelle erreur.

    - `-ansi` (nécessite `-pedantic`) spécifie la norme à laquelle le code doit se conformer. Actuellement, CPP connaît les normes C et C++ ; d'autres pourraient être ajoutés à l'avenir. L'option `-ansi` est équivalente à `-std=c89`, soit la norme "c89".

**Test 2 (Solution) :**
Après avoir analysé le Makefile et les différentes options j'ai compris que l'erreur venait 
- de l'option `-lm` a rajouter dans les fichiers nécessitant la librairie math.h
- de l'option `-ansi` puisque elle force le standard "c89" alors que la fonction `log2()` appartient au standard "c99"

ainsi il faut modifier le makefile de la manière suivante :

```
  GNU nano 4.8                                  Makefile                                             
mon_module_image.out: mon_module_image.o main.o
        gcc -o mon_module_image.out mon_module_image.o main.o -lm -std=c99

mon_module_image.o: gibert_module_image.c
        gcc -lm -std=c99 -c gibert_module_image.c -o mon_module_image.o -W -Wall -pedantic

main.o: main.c gibert_module_image.h
        gcc -o main.o -c main.c -std=c99 -W -Wall -pedantic

clean:
        rm -rf *.o
```
