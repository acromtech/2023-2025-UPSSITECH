```python
import scipy.io.wavfile as wav
import numpy as np
import matplotlib.pyplot as plt
from math import floor
import canaux as MEL
from math import pi
```

```python
# Récupère la fréquence du signal dans la variable "fs" et le signal dans la liste "signal"
fs, signal = wav.read('son.wav')

# Affiche la valeur de fs
print(fs)

# Affiche les 10 prmières valeur de signal
print(signal[0:10])
```

```python
# Affiche la durée du signal
print("la durée du signal est:",len(signal)/fs,"s")
```

```python
extrait=signal[47500:49547,1]
ham1024 = np.hamming(len(extrait))

# Limite la fenetre du graphique à la taille de l'extrait
fenetrage = np.multiply(extrait,ham1024)

#Détermine le cesptre de l'extrait
abs_fft = abs(np.fft.fft(fenetrage))
log_extrait = np.log(abs_fft)
cesptre = np.fft.fft(log_extrait)

# Divise en 3 champs -> subpolt
# Se positionne dans le premier champ graphique
plt.subplot(3,1,1)
# Affecte le titre "extrait" au graphique
plt.title("Extrait")
plt.xlabel("Temps[s]")
plt.ylabel("Amplitude[]")
# Affiche l'extrait du signal
plt.plot((np.arange(len(extrait))/fs),extrait)
# Ajuste la fenetre
plt.tight_layout()
plt.tight_layout()

# Se positionnne dans le second champ graphique
plt.subplot(3,1,2)
# Calcule le spectre le l'extrait
spectre_f = abs(np.fft.fft(extrait))
# Affecte le titre spectre au graphique
plt.title("Spectre")
plt.xlabel("Fréquence[Hz]")
plt.ylabel("Amplitude[]")

# Affiche les 512 premières valeurs du spectre ???
plt.plot((np.arange(len(spectre_f[:512]))/len(spectre_f[:512]))*fs,spectre_f[:512])
# Ajuste la fenetre
plt.tight_layout()
plt.tight_layout()

# Se positionnne dans le dernier champ graphique
plt.subplot(3,1,3)
# Affecte le titre cesptre au graphique
plt.title("Cesptre")
plt.xlabel("Qfréquence[]")
plt.ylabel("Amplitude[]")
# Affiche les 512 premières valeurs du cesptre
plt.plot((np.arange(len(extrait[:512]))/fs),cesptre[:512])
```

```python
#Donne la fréquence d'échantillonage du signal
fe = 8000

#(1/10)s la durée d'un silence soit 8000 impulsion/s*0,1s
dureeSilence = floor(fe*0.1)

#(1/2)s la durée d'une note soit 8000 impulsion/s*0,5s
dureeNote = floor(fe*0.5)

#Définition unitaire d'une note
note=np.arange(dureeNote)*2*pi/fe

#Définition d'un silence soit des zéros toute la durée d'un silence
silence = np.zeros(dureeSilence)

#Définition des différentes notes avec leurs fréquences respectives
do3 = np.sin(note*262)
re3 = np.sin(note*294)
mi3 = np.sin(note*330)

#Création de la partition de musique soit la concaténation des différentes notes avec leurs durées respectives intrinsèques
signal=np.concatenate((silence, do3, silence, do3, silence, do3, silence, re3, silence, mi3, mi3, silence, re3, re3, silence, do3, silence, mi3, silence, re3, silence,re3,silence,do3,do3,silence))

#Création d'un fichier .wav contenant la partition
wav.write('signal.wav',fe,signal)
```

```python
#Définition de la taille de la fenetre et du pas donné
taillefenetre = 1024
pas = 512

#Définition du nombre de colonne du graphique selon la taille de la fenetre et le pas
#Note : la fonction floor permet de récupérer seulement la valeur entière
nbcol=floor((len(signal)-taillefenetre)/pas)+1

#
Sp=np.zeros((floor(taillefenetre/2),nbcol))

#Pour chaque colonne on calcule la transposée du spectre 
for x in range (nbcol) :
    i=x*pas
    spectre=abs(np.fft.fft(signal[i:i+taillefenetre]))
    Sp[:,x]=np.transpose(spectre[:floor(taillefenetre/2)])
    
#On détermine la durée en seconde séparant chaque échantillons
duree=len(signal)/fe

#On détermine la fréquence maximale du signal
fmax=fe/2

#Paramètre le graphique du signal Sp
# Pour le paramètre "extent" [gauche, droite, bas, haut] :
#   Donne les coordonnées de données que l'image remplira. L'image est étirée individuellement le long de x et y pour remplir la boîte.
# Pour le paramètre "aspect" deux choix sont possible d'après la bibliothèque logiciellle de matplotlib
#  'equal' : garantit un rapport d'aspect de 1. Les pixels seront carrés (à moins que les tailles de pixel ne soient explicitement rendues non carrées dans les coordonnées de données à l'aide de extent ).
#  'auto' : les axes sont maintenus fixes et l'aspect est ajusté de manière à ce que les données tiennent dans les axes. En général, cela se traduira par des pixels non carrés.
# Pour le paramètre "origin" :
#  "upper" permet de placer l'index [0,0] du tableau dans le coin supérieur gauche des axes,
#  "lower" permet de placer l'index [0,0] du tableau dans le coin inférieur gauche des axes. 
#   Note : La convention (la valeur par défaut) 'upper' est généralement utilisée pour les matrices et les images.
plt.imshow(Sp,extent=[0,duree,0,fmax/1000],aspect="auto",origin="lower")

#Affiche le graphique généré
plt.show() 
```
