```python
#TP2
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
from math import floor
from math import pi
```

```python
fe = 8000
dureeSilence =floor(fe*0.1)
dureeNote = fe*1
note=np.arange(dureeNote)*2*pi/fe
silence = np.zeros(dureeSilence)
do3 = np.sin(note*262)
mi3 = np.sin(note*330)
sol3 = np.sin(note*392)
do4 = np.sin(note*523)
signal=np.concatenate((silence,do3,silence,mi3,silence,sol3,silence,do4,silence))
wav.write('signal.wav',fe,signal)
```

```python
taillefenetre = 1024
pas = 512
nbcol=floor((len(signal)-taillefenetre)/pas)+1
print(nbcol)
Sp=np.zeros((floor(taillefenetre/2),nbcol))

for x in range (nbcol) :
    i=x*pas
    spectre=abs(np.fft.fft(signal[i:i+taillefenetre]))
    Sp[:,x]=np.transpose(spectre[:floor(taillefenetre/2)])
duree=len(signal)/fe
fmax=fe/2
plt.imshow(Sp,extent=[0,duree,0,fmax],aspect="auto",origin="lower")
plt.show() 
```
